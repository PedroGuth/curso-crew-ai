### **üîç Criando a Ferramenta de Pesquisa de Tend√™ncias**

Vamos criar uma ferramenta que simula a pesquisa de tend√™ncias em tempo real. Esta ferramenta vai identificar t√≥picos quentes e oportunidades de conte√∫do.

- **Monitora** Google Trends e redes sociais
- **Identifica** t√≥picos em alta
- **Analisa** sazonalidade
- **Sugere** oportunidades de conte√∫do

#### **üîß Como nossa ferramenta vai funcionar:**
1. **Recebe** nicho/segmento
2. **Simula** busca de tend√™ncias
3. **Analisa** oportunidades
4. **Retorna** insights acion√°veis

````python
# üîç CRIANDO A FERRAMENTA DE PESQUISA DE TEND√äNCIAS
# Ferramenta que simula pesquisa de tend√™ncias em tempo real

class TrendResearchInput(BaseModel):
    """Input para pesquisa de tend√™ncias"""
    nicho: str = Field(description="Nicho ou segmento de mercado")
    periodo: str = Field(description="Per√≠odo de an√°lise (7d, 30d, 90d)")

class TrendResearchTool(BaseTool):
    name: str = "Pesquisador de Tend√™ncias"
    description: str = "Pesquisa tend√™ncias e t√≥picos quentes em tempo real"
    
    def _run(self, nicho: str, periodo: str = "30d") -> str:
        """
        Pesquisa tend√™ncias para um nicho espec√≠fico
        
        Args:
            nicho: Nicho ou segmento de mercado
            periodo: Per√≠odo de an√°lise
            
        Returns:
            An√°lise de tend√™ncias com oportunidades
        """
        
        # Simulando dados de tend√™ncias por nicho
        tendencias_data = {
            "tecnologia": {
                "tendencias_quentes": [
                    {"topico": "Intelig√™ncia Artificial", "crescimento": 85, "dificuldade": "M√©dia"},
                    {"topico": "Metaverso", "crescimento": 120, "dificuldade": "Alta"},
                    {"topico": "Criptomoedas", "crescimento": 65, "dificuldade": "M√©dia"},
                    {"topico": "Sustentabilidade Tech", "crescimento": 95, "dificuldade": "Baixa"}
                ],
                "hashtags_populares": ["#TechTrends", "#AI", "#Innovation", "#FutureTech"],
                "palavras_chave": ["tecnologia", "inova√ß√£o", "startup", "digital"]
            },
            "marketing": {
                "tendencias_quentes": [
                    {"topico": "Marketing Digital", "crescimento": 75, "dificuldade": "M√©dia"},
                    {"topico": "Influencer Marketing", "crescimento": 110, "dificuldade": "Alta"},
                    {"topico": "Marketing de Conte√∫do", "crescimento": 60, "dificuldade": "Baixa"},
                    {"topico": "Marketing Automation", "crescimento": 90, "dificuldade": "M√©dia"}
                ],
                "hashtags_populares": ["#MarketingDigital", "#GrowthHacking", "#ContentMarketing", "#SocialMedia"],
                "palavras_chave": ["marketing", "estrat√©gia", "crescimento", "vendas"]
            },
            "saude": {
                "tendencias_quentes": [
                    {"topico": "Sa√∫de Mental", "crescimento": 95, "dificuldade": "M√©dia"},
                    {"topico": "Fitness em Casa", "crescimento": 80, "dificuldade": "Baixa"},
                    {"topico": "Nutri√ß√£o Funcional", "crescimento": 70, "dificuldade": "M√©dia"},
                    {"topico": "Telemedicina", "crescimento": 100, "dificuldade": "Alta"}
                ],
                "hashtags_populares": ["#SaudeMental", "#Fitness", "#Nutricao", "#BemEstar"],
                "palavras_chave": ["sa√∫de", "bem-estar", "fitness", "nutri√ß√£o"]
            }
        }
        
        # Obtendo dados do nicho
        nicho_lower = nicho.lower()
        if nicho_lower not in tendencias_data:
            return f"Nicho '{nicho}' n√£o encontrado. Nichos dispon√≠veis: {list(tendencias_data.keys())}"
        
        dados_nicho = tendencias_data[nicho_lower]
        
        # Criando an√°lise de tend√™ncias
        relatorio = f"""
ÔøΩÔøΩ PESQUISA DE TEND√äNCIAS - {nicho.upper()}
==========================================

ÔøΩÔøΩ PER√çODO ANALISADO: {periodo}
üéØ NICHO: {nicho}

ÔøΩÔøΩ TEND√äNCIAS QUENTES:
"""
        
        for i, tendencia in enumerate(dados_nicho["tendencias_quentes"], 1):
            relatorio += f"""

{i}. {tendencia['topico']}
   üìä Crescimento: +{tendencia['crescimento']}%
   ÔøΩÔøΩ Dificuldade: {tendencia['dificuldade']}
   ÔøΩÔøΩ Oportunidade: {'Alta' if tendencia['crescimento'] > 80 else 'M√©dia' if tendencia['crescimento'] > 60 else 'Baixa'}
"""
        
        relatorio += f"""

üè∑Ô∏è HASHTAGS POPULARES:
"""
        
        for hashtag in dados_nicho["hashtags_populares"]:
            relatorio += f"‚Ä¢ {hashtag}\n"
        
        relatorio += f"""

ÔøΩÔøΩ PALAVRAS-CHAVE SUGERIDAS:
"""
        
        for palavra in dados_nicho["palavras_chave"]:
            relatorio += f"‚Ä¢ {palavra}\n"
        
        relatorio += f"""

üí° RECOMENDA√á√ïES DE CONTE√öDO:
‚Ä¢ Foque em t√≥picos com crescimento > 80%
‚Ä¢ Use hashtags populares estrategicamente
‚Ä¢ Crie conte√∫do educativo sobre tend√™ncias
‚Ä¢ Mantenha-se atualizado com novidades do setor

‚úÖ Status: An√°lise de tend√™ncias conclu√≠da!
üìä Total de oportunidades identificadas: {len(dados_nicho['tendencias_quentes'])}
"""
        
        return relatorio

print("‚úÖ Pesquisador de Tend√™ncias criado com sucesso!")
print("üîß Testando a ferramenta...")

# Testando nossa ferramenta
pesquisador_tendencias = TrendResearchTool()
resultado_teste = pesquisador_tendencias._run(
    nicho="tecnologia",
    periodo="30d"
)

print("\nÔøΩÔøΩ RESULTADO DO TESTE:")
print(resultado_teste)
print("=" * 70)
````

### **üìä Criando a Ferramenta de An√°lise de SEO**

Agora vamos criar uma ferramenta que analisa e otimiza conte√∫do para SEO. Esta ferramenta vai sugerir palavras-chave e otimiza√ß√µes.

#### **üéØ O que faz um Analista de SEO?**
- **Sugere** palavras-chave
- **Analisa** densidade de termos
- **Otimiza** para busca org√¢nica
- **Monitora** rankings

#### **üîß Como nossa ferramenta vai funcionar:**
1. **Recebe** t√≥pico e conte√∫do
2. **Analisa** palavras-chave
3. **Sugere** otimiza√ß√µes
4. **Retorna** score de SEO

````python
# üìä CRIANDO A FERRAMENTA DE AN√ÅLISE DE SEO
# Ferramenta para otimiza√ß√£o de conte√∫do para busca org√¢nica

class SEOAnalysisInput(BaseModel):
    """Input para an√°lise de SEO"""
    topico: str = Field(description="T√≥pico principal do conte√∫do")
    conteudo: str = Field(description="Conte√∫do para an√°lise")
    palavras_chave: list = Field(description="Palavras-chave alvo")

class SEOAnalysisTool(BaseTool):
    name: str = "Analisador de SEO"
    description: str = "Analisa e otimiza conte√∫do para SEO"
    
    def _run(self, topico: str, conteudo: str, palavras_chave: list) -> str:
        """
        Analisa conte√∫do para otimiza√ß√£o SEO
        
        Args:
            topico: T√≥pico principal
            conteudo: Conte√∫do para an√°lise
            palavras_chave: Palavras-chave alvo
            
        Returns:
            An√°lise SEO completa com recomenda√ß√µes
        """
        
        # Simulando an√°lise SEO
        conteudo_lower = conteudo.lower()
        palavras_conteudo = conteudo_lower.split()
        total_palavras = len(palavras_conteudo)
        
        # Analisando densidade de palavras-chave
        densidades = {}
        for palavra in palavras_chave:
            palavra_lower = palavra.lower()
            ocorrencias = conteudo_lower.count(palavra_lower)
            densidade = (ocorrencias / total_palavras) * 100 if total_palavras > 0 else 0
            densidades[palavra] = {
                "ocorrencias": ocorrencias,
                "densidade": densidade,
                "status": "√ìtima" if 1 <= densidade <= 3 else "Baixa" if densidade < 1 else "Alta"
            }
        
        # Calculando score geral de SEO
        score_seo = 0
        criterios = []
        
        # Crit√©rio 1: Densidade de palavras-chave
        densidade_media = np.mean([d["densidade"] for d in densidades.values()])
        if 1 <= densidade_media <= 3:
            score_seo += 25
            criterios.append("‚úÖ Densidade de palavras-chave ideal")
        else:
            criterios.append("‚ö†Ô∏è Densidade de palavras-chave precisa de ajuste")
        
        # Crit√©rio 2: Tamanho do conte√∫do
        if total_palavras >= 300:
            score_seo += 20
            criterios.append("‚úÖ Tamanho do conte√∫do adequado")
        else:
            criterios.append("‚ö†Ô∏è Conte√∫do muito curto para SEO")
        
        # Crit√©rio 3: Presen√ßa de palavras-chave no t√≠tulo
        if any(palavra.lower() in topico.lower() for palavra in palavras_chave):
            score_seo += 25
            criterios.append("‚úÖ Palavra-chave no t√≠tulo")
        else:
            criterios.append("‚ö†Ô∏è Palavra-chave n√£o encontrada no t√≠tulo")
        
        # Crit√©rio 4: Estrutura do conte√∫do
        if "#" in conteudo or "**" in conteudo:
            score_seo += 15
            criterios.append("‚úÖ Estrutura de conte√∫do adequada")
        else:
            criterios.append("‚ö†Ô∏è Estrutura de conte√∫do pode ser melhorada")
        
        # Crit√©rio 5: Links internos/externos
        if "http" in conteudo:
            score_seo += 15
            criterios.append("‚úÖ Links presentes no conte√∫do")
        else:
            criterios.append("‚ö†Ô∏è Considere adicionar links relevantes")
        
        # Classificando score
        if score_seo >= 80:
            classificacao = "EXCELENTE"
            recomendacao = "Conte√∫do muito bem otimizado para SEO!"
        elif score_seo >= 60:
            classificacao = "BOM"
            recomendacao = "Conte√∫do bem otimizado, com pequenos ajustes pode melhorar."
        elif score_seo >= 40:
            classificacao = "REGULAR"
            recomendacao = "Conte√∫do precisa de otimiza√ß√µes para melhorar SEO."
        else:
            classificacao = "PRECISA MELHORAR"
            recomendacao = "Conte√∫do precisa de otimiza√ß√µes significativas."
        
        # Criando relat√≥rio
        relatorio = f"""
ÔøΩÔøΩ AN√ÅLISE SEO - {topico.upper()}
==============================

ÔøΩÔøΩ T√ìPICO: {topico}
üìù TAMANHO: {total_palavras} palavras
üîë PALAVRAS-CHAVE: {', '.join(palavras_chave)}

üèÜ SCORE SEO: {score_seo}/100 ({classificacao})

üìà AN√ÅLISE DE DENSIDADE:
"""
        
        for palavra, dados in densidades.items():
            relatorio += f"‚Ä¢ {palavra}: {dados['ocorrencias']} ocorr√™ncias ({dados['densidade']:.2f}%) - {dados['status']}\n"
        
        relatorio += f"""

‚úÖ CRIT√âRIOS AVALIADOS:
"""
        
        for criterio in criterios:
            relatorio += f"{criterio}\n"
        
        relatorio += f"""

üí° RECOMENDA√á√ïES:
{recomendacao}

üîß OTIMIZA√á√ïES SUGERIDAS:
‚Ä¢ Ajuste densidade de palavras-chave para 1-3%
‚Ä¢ Adicione subt√≠tulos com H2 e H3
‚Ä¢ Inclua links internos e externos relevantes
‚Ä¢ Otimize meta descri√ß√£o
‚Ä¢ Adicione imagens com alt text

‚úÖ Status: An√°lise SEO conclu√≠da!
"""
        
        return relatorio

print("‚úÖ Analisador de SEO criado com sucesso!")
print("üîß Testando a ferramenta...")

# Testando nossa ferramenta
analisador_seo = SEOAnalysisTool()
resultado_teste = analisador_seo._run(
    topico="Intelig√™ncia Artificial em 2024",
    conteudo="A Intelig√™ncia Artificial est√° revolucionando diversos setores. A IA tem se mostrado fundamental para o futuro da tecnologia.",
    palavras_chave=["Intelig√™ncia Artificial", "IA", "tecnologia"]
)

print("\nÔøΩÔøΩ RESULTADO DO TESTE:")
print(resultado_teste)
print("=" * 70)
````

### **üìà Criando a Ferramenta de An√°lise de Performance**

Agora vamos criar uma ferramenta que monitora e analisa a performance do conte√∫do publicado. Esta ferramenta vai gerar insights sobre engajamento e ROI.

#### **üéØ O que faz um Analista de Performance?**

- **Analisa** convers√µes
- **Calcula** ROI de conte√∫do

#### **üîß Como nossa ferramenta vai funcionar:**

2. **Calcula** m√©tricas importantes

````python
# üìà CRIANDO A FERRAMENTA DE AN√ÅLISE DE PERFORMANCE
# Ferramenta para monitorar e analisar performance de conte√∫do

class PerformanceAnalysisInput(BaseModel):
    """Input para an√°lise de performance"""
    titulo: str = Field(description="T√≠tulo do conte√∫do")
    plataforma: str = Field(description="Plataforma de publica√ß√£o")
    periodo: str = Field(description="Per√≠odo de an√°lise")

class PerformanceAnalysisTool(BaseTool):
    name: str = "Analisador de Performance"
    description: str = "Analisa performance e ROI de conte√∫do publicado"
    
    def _run(self, titulo: str, plataforma: str, periodo: str = "7d") -> str:
        """
        Analisa performance de conte√∫do publicado
        
        Args:
            titulo: T√≠tulo do conte√∫do
            plataforma: Plataforma de publica√ß√£o
            periodo: Per√≠odo de an√°lise
            
        Returns:
            An√°lise completa de performance
        """
        
        # Simulando dados de performance
        dados_performance = {
            "blog": {
                "visualizacoes": random.randint(1000, 10000),
                "tempo_medio": random.randint(2, 8),
                "taxa_rejeicao": random.uniform(20, 60),
                "comentarios": random.randint(5, 50),
                "compartilhamentos": random.randint(10, 200),
                "conversoes": random.randint(5, 100)
            },
            "linkedin": {
                "visualizacoes": random.randint(500, 5000),
                "likes": random.randint(20, 200),
                "comentarios": random.randint(5, 30),
                "compartilhamentos": random.randint(2, 50),
                "cliques": random.randint(50, 500),
                "conversoes": random.randint(2, 20)
            },
            "instagram": {
                "alcance": random.randint(2000, 15000),
                "impressoes": random.randint(3000, 20000),
                "likes": random.randint(100, 1000),
                "comentarios": random.randint(10, 100),
                "salvos": random.randint(20, 200),
                "cliques_perfil": random.randint(50, 300)
            }
        }
        
        # Obtendo dados da plataforma
        plataforma_lower = plataforma.lower()
        if plataforma_lower not in dados_performance:
            return f"Plataforma '{plataforma}' n√£o suportada. Plataformas: {list(dados_performance.keys())}"
        
        dados = dados_performance[plataforma_lower]
        
        # Calculando m√©tricas derivadas
        if plataforma_lower == "blog":
            taxa_engajamento = ((dados["comentarios"] + dados["compartilhamentos"]) / dados["visualizacoes"]) * 100
            roi_conteudo = (dados["conversoes"] * 100) / dados["visualizacoes"]  # Simulando valor por convers√£o
        elif plataforma_lower == "linkedin":
            taxa_engajamento = ((dados["likes"] + dados["comentarios"] + dados["compartilhamentos"]) / dados["visualizacoes"]) * 100
            roi_conteudo = (dados["conversoes"] * 50) / dados["visualizacoes"]  # Simulando valor por convers√£o
        else:  # instagram
            taxa_engajamento = ((dados["likes"] + dados["comentarios"]) / dados["alcance"]) * 100
            roi_conteudo = (dados["cliques_perfil"] * 2) / dados["alcance"]  # Simulando valor por clique
        
        # Classificando performance
        if taxa_engajamento >= 5:
            classificacao_engajamento = "EXCELENTE"
        elif taxa_engajamento >= 3:
            classificacao_engajamento = "BOM"
        elif taxa_engajamento >= 1:
            classificacao_engajamento = "REGULAR"
        else:
            classificacao_engajamento = "BAIXO"
        
        # Criando relat√≥rio
        relatorio = f"""
ÔøΩÔøΩ AN√ÅLISE DE PERFORMANCE - {titulo.upper()}
==========================================

ÔøΩÔøΩ PLATAFORMA: {plataforma.upper()}
üìÖ PER√çODO: {periodo}
ÔøΩÔøΩ T√çTULO: {titulo}

üìä M√âTRICAS PRINCIPAIS:
"""
        
        for metrica, valor in dados.items():
            if isinstance(valor, float):
                relatorio += f"‚Ä¢ {metrica.replace('_', ' ').title()}: {valor:.1f}\n"
            else:
                relatorio += f"‚Ä¢ {metrica.replace('_', ' ').title()}: {valor:,}\n"
        
        relatorio += f"""

ÔøΩÔøΩ M√âTRICAS DERIVADAS:
‚Ä¢ Taxa de Engajamento: {taxa_engajamento:.2f}% ({classificacao_engajamento})
‚Ä¢ ROI do Conte√∫do: {roi_conteudo:.2f}%

üèÜ CLASSIFICA√á√ÉO GERAL:
‚Ä¢ Engajamento: {classificacao_engajamento}
‚Ä¢ Alcance: {'Alto' if max(dados.values()) > 5000 else 'M√©dio' if max(dados.values()) > 1000 else 'Baixo'}
‚Ä¢ Convers√£o: {'Alta' if 'conversoes' in dados and dados['conversoes'] > 20 else 'M√©dia' if 'conversoes' in dados and dados['conversoes'] > 5 else 'Baixa'}

üí° INSIGHTS:
‚Ä¢ {'Excelente engajamento! Mantenha a estrat√©gia.' if taxa_engajamento >= 5 else 'Bom engajamento, h√° espa√ßo para melhorar.' if taxa_engajamento >= 3 else 'Engajamento baixo, revise a estrat√©gia.'}
‚Ä¢ {'Alto ROI! Conte√∫do muito eficiente.' if roi_conteudo >= 5 else 'ROI positivo, pode otimizar mais.' if roi_conteudo >= 2 else 'ROI baixo, precisa de otimiza√ß√£o.'}

üîß RECOMENDA√á√ïES:
‚Ä¢ {'Continue com a estrat√©gia atual' if taxa_engajamento >= 5 else 'Teste diferentes hor√°rios de publica√ß√£o' if taxa_engajamento >= 3 else 'Revise o tipo de conte√∫do e p√∫blico-alvo'}
‚Ä¢ {'Aumente a frequ√™ncia de publica√ß√£o' if roi_conteudo >= 5 else 'Otimize call-to-actions' if roi_conteudo >= 2 else 'Reformule a estrat√©gia de conte√∫do'}

‚úÖ Status: An√°lise de performance conclu√≠da!
ÔøΩÔøΩ Per√≠odo analisado: {periodo}
"""
        
        return relatorio

print("‚úÖ Analisador de Performance criado com sucesso!")
print("üîß Testando a ferramenta...")

# Testando nossa ferramenta
analisador_performance = PerformanceAnalysisTool()
resultado_teste = analisador_performance._run(
    titulo="Guia Completo de IA em 2024",
    plataforma="blog",
    periodo="7d"
)

print("\nÔøΩÔøΩ RESULTADO DO TESTE:")
print(resultado_teste)
print("=" * 70)
````

### **üìã Criando Tarefas para o Sistema Completo**

Agora vamos criar tarefas que formam um fluxo completo de gest√£o de conte√∫do. Cada tarefa vai ser executada por um especialista espec√≠fico.

#### **üéØ Fluxo de Tarefas:**

1. **Pesquisa de Mercado** - Identificar oportunidades
2. **An√°lise de SEO** - Otimizar para busca
3. **Cria√ß√£o de Conte√∫do** - Escrever artigo
4. **Design Visual** - Criar visualiza√ß√µes
5. **Revis√£o de Qualidade** - Garantir excel√™ncia
6. **Publica√ß√£o** - Coordenar lan√ßamento
7. **Monitoramento** - Analisar performance
8. **Relat√≥rio Final** - Consolidar resultados

#### **üîÑ Depend√™ncias:**
Cada tarefa depende das anteriores, criando um fluxo sequencial mas otimizado!

````python
# üìã CRIANDO TAREFAS PARA O SISTEMA COMPLETO
# Tarefas que formam um fluxo completo de gest√£o de conte√∫do

print("üìã Criando tarefas para o sistema completo de gest√£o de conte√∫do...")

# Tarefa 1: Pesquisa de Mercado (Pesquisador)
tarefa_pesquisa = Task(
    description="""
    Pesquise tend√™ncias e oportunidades de conte√∫do para o nicho de tecnologia.
    
    Use sua ferramenta de pesquisa de tend√™ncias para:
    - Identificar t√≥picos quentes em tecnologia
    - Analisar crescimento de tend√™ncias
    - Sugerir oportunidades de conte√∫do
    - Identificar palavras-chave populares
    
    Foque em tend√™ncias que tenham potencial de engajamento e convers√£o.
    """,
    agent=pesquisador_mercado,
    expected_output="Relat√≥rio completo de tend√™ncias e oportunidades de conte√∫do"
)

# Tarefa 2: An√°lise de SEO (Analista SEO)
tarefa_seo = Task(
    description="""
    Analise e otimize o conte√∫do para SEO baseado nas tend√™ncias identificadas.
    
    Use sua ferramenta de an√°lise SEO para:
    - Sugerir palavras-chave otimizadas
    - Analisar densidade de termos
    - Otimizar estrutura do conte√∫do
    - Recomendar melhorias t√©cnicas
    
    Baseie-se nas tend√™ncias identificadas pelo pesquisador.
    """,
    agent=analista_seo,
    expected_output="An√°lise SEO completa com recomenda√ß√µes de otimiza√ß√£o",
    context=[tarefa_pesquisa]  # Depende da pesquisa
)

# Tarefa 3: Cria√ß√£o de Conte√∫do (Redator)
tarefa_criacao = Task(
    description="""
    Crie um artigo completo sobre uma das tend√™ncias identificadas.
    
    Com base na pesquisa e an√°lise SEO:
    - Escreva um artigo de 1500-2000 palavras
    - Use as palavras-chave sugeridas
    - Crie uma narrativa envolvente
    - Inclua exemplos pr√°ticos
    - Adicione call-to-actions estrat√©gicos
    
    O artigo deve ser informativo, envolvente e otimizado para convers√£o.
    """,
    agent=redator_criativo,
    expected_output="Artigo completo otimizado para SEO e convers√£o",
    context=[tarefa_pesquisa, tarefa_seo]  # Depende da pesquisa e SEO
)

# Tarefa 4: Design Visual (Designer)
tarefa_design = Task(
    description="""
    Crie visualiza√ß√µes complementares para o artigo.
    
    Com base no conte√∫do criado:
    - Crie infogr√°ficos explicativos
    - Desenvolva gr√°ficos de dados
    - Produza imagens para redes sociais
    - Otimize para diferentes plataformas
    
    As visualiza√ß√µes devem complementar e fortalecer o conte√∫do escrito.
    """,
    agent=designer_conteudo,
    expected_output="Pacote completo de visualiza√ß√µes para o artigo",
    context=[tarefa_criacao]  # Depende da cria√ß√£o do conte√∫do
)

# Tarefa 5: Revis√£o de Qualidade (Revisor)
tarefa_revisao = Task(
    description="""
    Revise todo o conte√∫do criado para garantir qualidade e precis√£o.
    
    Revise:
    - Precis√£o das informa√ß√µes
    - Clareza da escrita
        agent=revisor_qualidade,
    expected_output="Conte√∫do revisado e aprovado para publica√ß√£o",
    context=[tarefa_criacao, tarefa_design]  # Depende da cria√ß√£o e design
)

# Tarefa 6: Publica√ß√£o (Gerente)
tarefa_publicacao = Task(
    description="""
    Coordene o lan√ßamento do conte√∫do em m√∫ltiplas plataformas.
    
    Organize:
    - Cronograma de publica√ß√£o
    - Adapta√ß√£o para diferentes plataformas
    - Estrat√©gia de distribui√ß√£o
    - Monitoramento inicial
    
    Garanta que o conte√∫do seja publicado no momento ideal.
    """,
    agent=gerente_publicacao,
    expected_output="Plano de publica√ß√£o e distribui√ß√£o executado",
    context=[tarefa_revisao]  # Depende da revis√£o
)

# Tarefa 7: Monitoramento (Analista Performance)
tarefa_monitoramento = Task(
    description="""
    Monitore a performance do conte√∫do publicado.
    
    Use sua ferramenta de an√°lise para:
    - Acompanhar m√©tricas de engajamento
    - Analisar convers√µes
    - Identificar oportunidades de otimiza√ß√£o
    - Gerar insights acion√°veis
    
    Forne√ßa dados para tomada de decis√£o estrat√©gica.
    """,
    agent=analista_performance,
    expected_output="Relat√≥rio completo de performance do conte√∫do",
    context=[tarefa_publicacao]  # Depende da publica√ß√£o
)

# Tarefa 8: Relat√≥rio Final (Diretor)
tarefa_relatorio_final = Task(
    description="""
    Crie um relat√≥rio final consolidando todos os resultados.
    
    Integre:
    - Resultados da pesquisa de mercado
    - Performance do conte√∫do
    - ROI da campanha
    - Li√ß√µes aprendidas
    - Recomenda√ß√µes futuras
    
    Este relat√≥rio deve orientar estrat√©gias futuras.
    """,
    agent=diretor_marketing,
    expected_output="Relat√≥rio estrat√©gico completo com recomenda√ß√µes",
    context=[tarefa_pesquisa, tarefa_seo, tarefa_criacao, tarefa_design, tarefa_revisao, tarefa_publicacao, tarefa_monitoramento]  # Depende de todas as anteriores
)

print("‚úÖ Tarefas para o sistema completo criadas com sucesso!")
print(f"   Tarefa 1: Pesquisa de Mercado")
print(f"   Tarefa 2: An√°lise de SEO (depende da 1)")
print(f"   Tarefa 3: Cria√ß√£o de Conte√∫do (depende da 1 e 2)")
print(f"   Tarefa 4: Design Visual (depende da 3)")
print(f"   Tarefa 5: Revis√£o de Qualidade (depende da 3 e 4)")
print(f"   Tarefa 6: Publica√ß√£o (depende da 5)")
print(f"   Tarefa 7: Monitoramento (depende da 6)")
print(f"   Tarefa 8: Relat√≥rio Final (depende de todas)")
print("=" * 70)
````
