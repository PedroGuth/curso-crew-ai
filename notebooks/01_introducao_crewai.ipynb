## **ðŸ—ï¸ Aula 1.3: Como Funciona a Arquitetura do CrewAI?**

### **ðŸ”§ Componentes Fundamentais**

Vamos entender cada peÃ§a do quebra-cabeÃ§a:

#### **ðŸ‘¤ 1. Agent (Agente)**
**O que Ã©**: Uma IA especializada com um papel especÃ­fico

**CaracterÃ­sticas**:
- **Role**: FunÃ§Ã£o especÃ­fica (ex: "Analista de Dados")
- **Goal**: Objetivo principal do agente
- **Backstory**: Contexto e experiÃªncia do agente
- **Tools**: Ferramentas que o agente pode usar
- **LLM**: Modelo de linguagem que "pensa" pelo agente

**Analogia**: Como definir um funcionÃ¡rio - nome, cargo, experiÃªncia, ferramentas

#### **ðŸ“‹ 2. Task (Tarefa)**
**O que Ã©**: Um trabalho especÃ­fico que precisa ser feito

**CaracterÃ­sticas**:
- **Description**: O que precisa ser feito
- **Agent**: Quem vai fazer
- **Expected Output**: O que esperamos receber

**Analogia**: Como uma tarefa no Trello - descriÃ§Ã£o, responsÃ¡vel, prazo, dependÃªncias

#### **ðŸ”§ 3. Tool (Ferramenta)**
**O que Ã©**: Uma ferramenta que o agente pode usar

**Exemplos**:
- **Web Search**: Buscar informaÃ§Ãµes na internet
- **File Reader**: Ler arquivos
- **Calculator**: Fazer cÃ¡lculos
- **Custom Tools**: Ferramentas que vocÃª cria

**Analogia**: Como a caixa de ferramentas de um pedreiro - martelo, chave de fenda, etc.

#### **ðŸ”„ 4. Process (Processo)**
**O que Ã©**: Como os agentes trabalham juntos

**Tipos**:
- **Sequential**: Um depois do outro (como uma linha de montagem)
- **Hierarchical**: Com hierarquia (gerente â†’ funcionÃ¡rios)
- **Collaborative**: Trabalhando juntos

**Analogia**: Como o fluxo de trabalho em uma empresa

#### **ðŸ‘¥ 5. Crew (Equipe)**
**O que Ã©**: O grupo de agentes trabalhando juntos

**CaracterÃ­sticas**:
- **Agents**: Lista de agentes na equipe
- **Tasks**: Lista de tarefas a serem feitas
- **Process**: Como eles vÃ£o trabalhar
- **Goal**: Objetivo final da equipe

**Analogia**: Como uma equipe de projeto - pessoas, tarefas, metodologia, objetivo

### **ðŸ”„ Fluxo de Trabalho TÃ­pico**

```
1. Definir o problema/objetivo
```

## **ðŸŽ¯ Aula 1.4: Primeiro Exemplo PrÃ¡tico - Nosso EscritÃ³rio Virtual**

Agora vamos colocar a mÃ£o na massa! Vamos criar nosso primeiro sistema CrewAI - um escritÃ³rio virtual com 3 agentes bÃ¡sicos.

### **ðŸ¢ Nosso Projeto: EscritÃ³rio de Marketing Digital**

Vamos criar um sistema que:
- **Pesquisa** tendÃªncias de mercado
- **Analisa** dados e insights
- **Cria** conteÃºdo baseado nas descobertas

### **ðŸ‘¥ Nossos Agentes:**
1. **Pesquisador de Mercado** - Encontra informaÃ§Ãµes
2. **Analista de Dados** - Processa e interpreta
3. **Redator Criativo** - Cria conteÃºdo

Vamos comeÃ§ar criando nossos primeiros agentes!

````python
# ðŸ¢ CRIANDO NOSSO PRIMEIRO SISTEMA CREWAI
# Vamos criar um escritÃ³rio virtual com 3 agentes bÃ¡sicos

# Importando as bibliotecas necessÃ¡rias
from crewai import Agent, Task, Crew, Process
from langchain_openai import ChatOpenAI
from langchain.llms import HuggingFaceHub
import os

print("ðŸš€ Iniciando criaÃ§Ã£o do nosso primeiro sistema CrewAI!")
print("=" * 60)
````

### **ðŸ”§ Configurando o LLM**

Primeiro, precisamos configurar o modelo de linguagem que nossos agentes vÃ£o usar. Ã‰ como escolher o "cÃ©rebro" que todos os funcionÃ¡rios vÃ£o compartilhar.

````python
# ðŸ§  CONFIGURANDO O LLM PARA NOSSOS AGENTES
# Vamos usar a funÃ§Ã£o que criamos no mÃ³dulo anterior

def get_llm_colab():
    """Retorna o melhor LLM disponÃ­vel no Colab"""
    
    # Tentativa 1: OpenAI
    try:
        api_key = os.getenv("OPENAI_API_KEY")
        if api_key:
            print("âœ… Usando OpenAI GPT-3.5-turbo")
            return ChatOpenAI(
                model="gpt-3.5-turbo",
                temperature=0.7,
                api_key=api_key
            )
    except:
        pass
    
    # Tentativa 2: Hugging Face
    try:
        token = os.getenv("HUGGINGFACEHUB_API_TOKEN")
        if token:
            print("âœ… Usando Hugging Face (gratuito)")
            return HuggingFaceHub(
                repo_id="google/flan-t5-base",
                model_kwargs={
                    "temperature": 0.7, 
                    "max_length": 512
                },
                huggingfacehub_api_token=token
            )
    except:
        pass
    
    print("âš ï¸ Usando modelo padrÃ£o")
    return None

# Configurando o LLM
llm = get_llm_colab()

if llm:
    print(f"ðŸŽ‰ LLM configurado: {type(llm).__name__}")
else:
    print("âš ï¸ LLM nÃ£o configurado - alguns recursos podem nÃ£o funcionar")
````

### **ðŸ‘¤ Criando Nosso Primeiro Agente - O Pesquisador**

Agora vamos criar nosso primeiro agente! Vamos comeÃ§ar com o **Pesquisador de Mercado**. Ã‰ como contratar o primeiro funcionÃ¡rio do nosso escritÃ³rio.

#### **ðŸ” O que faz um Pesquisador de Mercado?**
- **Busca** informaÃ§Ãµes sobre tendÃªncias
- **Coleta** dados de diferentes fontes
- **Identifica** oportunidades de mercado
- **Analisa** concorrÃªncia

#### **ðŸŽ¯ Como definimos um agente?**
Cada agente precisa de:
- **Role**: Qual Ã© a funÃ§Ã£o dele
- **Goal**: O que ele quer alcanÃ§ar
- **Backstory**: Sua experiÃªncia e contexto
- **Tools**: Que ferramentas ele pode usar
- **LLM**: Como ele "pensa"

````python
# ðŸ‘¤ CRIANDO O PESQUISADOR DE MERCADO
# Nosso primeiro agente - especialista em encontrar informaÃ§Ãµes

print("ðŸ” Criando o Pesquisador de Mercado...")

# Criando o agente pesquisador
pesquisador = Agent(
    role="Pesquisador de Mercado",
    goal="Encontrar as melhores informaÃ§Ãµes e tendÃªncias de mercado para criar estratÃ©gias eficazes",
    backstory="""
    VocÃª Ã© um pesquisador de mercado experiente com mais de 10 anos de experiÃªncia.
    VocÃª tem um olhar afiado para identificar tendÃªncias emergentes e oportunidades de negÃ³cio.
    Sua especialidade Ã© coletar dados de mÃºltiplas fontes e transformÃ¡-los em insights acionÃ¡veis.
    VocÃª sempre busca informaÃ§Ãµes atualizadas e confiÃ¡veis.
    """,
    verbose=True,  # Mostra o que o agente estÃ¡ pensando
    allow_delegation=False,  # NÃ£o delega tarefas para outros
    llm=llm  # Usando o LLM que configuramos
)

print("âœ… Pesquisador de Mercado criado com sucesso!")
print(f"   Role: {pesquisador.role}")
print(f"   Goal: {pesquisador.goal}")
print("=" * 60)
````

### **ðŸ“Š Criando o Segundo Agente - O Analista**

Agora vamos criar o **Analista de Dados**. Este agente vai pegar as informaÃ§Ãµes que o pesquisador encontrou e transformÃ¡-las em insights acionÃ¡veis.

#### **ðŸ” O que faz um Analista de Dados?**
- **Processa** dados brutos
- **Identifica** padrÃµes e tendÃªncias
- **Cria** visualizaÃ§Ãµes e relatÃ³rios
- **Extrai** insights acionÃ¡veis

#### **ðŸŽ¯ Por que precisamos de um analista separado?**
Ã‰ como ter um **pesquisador** que encontra as informaÃ§Ãµes e um **analista** que entende o que elas significam. SÃ£o especialidades diferentes!

````python
# ðŸ“Š CRIANDO O ANALISTA DE DADOS
# Nosso segundo agente - especialista em processar e interpretar dados

print("ðŸ“Š Criando o Analista de Dados...")

# Criando o agente analista
analista = Agent(
    role="Analista de Dados",
    goal="Transformar dados brutos em insights acionÃ¡veis e recomendaÃ§Ãµes estratÃ©gicas",
    backstory="""
    VocÃª Ã© um analista de dados sÃªnior com expertise em business intelligence.
    VocÃª tem habilidade excepcional para identificar padrÃµes em dados complexos.
    Sua especialidade Ã© transformar informaÃ§Ãµes tÃ©cnicas em recomendaÃ§Ãµes prÃ¡ticas.
    VocÃª sempre busca a histÃ³ria por trÃ¡s dos nÃºmeros.
    """,
    verbose=True,
    allow_delegation=False,
    llm=llm
)

print("âœ… Analista de Dados criado com sucesso!")
print(f"   Role: {analista.role}")
print(f"   Goal: {analista.goal}")
print("=" * 60)
````

### **âœï¸ Criando o Terceiro Agente - O Redator**

Por fim, vamos criar o **Redator Criativo**. Este agente vai pegar os insights do analista e transformÃ¡-los em conteÃºdo atrativo e persuasivo.

#### **ðŸ” O que faz um Redator Criativo?**
- **Cria** conteÃºdo atrativo
- **Adapta** linguagem para o pÃºblico
- **Estrutura** informaÃ§Ãµes de forma clara
- **Persuade** atravÃ©s da escrita

#### **ðŸŽ¯ O fluxo completo:**
**Pesquisador** â†’ **Analista** â†’ **Redator**

Ã‰ como uma linha de produÃ§Ã£o onde cada um faz sua parte!

````python
# âœï¸ CRIANDO O REDATOR CRIATIVO
# Nosso terceiro agente - especialista em criar conteÃºdo atrativo

print("âœï¸ Criando o Redator Criativo...")

# Criando o agente redator
redator = Agent(
    role="Redator Criativo",
    goal="Criar conteÃºdo atrativo e persuasivo baseado em insights de dados",
    backstory="""
    VocÃª Ã© um redator criativo experiente com talento para transformar dados complexos em histÃ³rias envolventes.
    VocÃª tem expertise em marketing digital e sabe como adaptar a linguagem para diferentes pÃºblicos.
    Sua especialidade Ã© criar conteÃºdo que nÃ£o apenas informa, mas tambÃ©m engaja e persuade.
    VocÃª sempre busca o Ã¢ngulo mais interessante para contar a histÃ³ria.
    """,
    verbose=True,
    allow_delegation=False,
    llm=llm
)

print("âœ… Redator Criativo criado com sucesso!")
print(f"   Role: {redator.role}")
print(f"   Goal: {redator.goal}")
print("=" * 60)
````

### **ðŸŽ¯ Criando Nossas Primeiras Tarefas**

Agora que temos nossos agentes, vamos criar as tarefas que cada um vai executar. Ã‰ como definir o que cada funcionÃ¡rio vai fazer no projeto.

#### **ðŸ“‹ O que Ã© uma Task?**
Uma **Task** Ã© um trabalho especÃ­fico que um agente precisa fazer. Ã‰ como uma tarefa no Trello ou Asana:
- **DescriÃ§Ã£o** clara do que precisa ser feito
- **ResponsÃ¡vel** (qual agente vai fazer)
- **Resultado esperado** (o que queremos receber)
- **DependÃªncias** (o que precisa ser feito antes)

#### **ðŸ”„ Nosso Fluxo de Tarefas:**
1. **Pesquisador** busca informaÃ§Ãµes sobre IA
2. **Analista** processa e interpreta os dados
3. **Redator** cria conteÃºdo baseado nos insights

````python
# ðŸ“‹ CRIANDO NOSSAS PRIMEIRAS TAREFAS
# Vamos definir o que cada agente vai fazer

print("ðŸ“‹ Criando as tarefas para nossos agentes...")

# Tarefa 1: Pesquisa de Mercado
tarefa_pesquisa = Task(
    description="""
    Pesquise as principais tendÃªncias de InteligÃªncia Artificial em 2024.
    
    Foque em:
    - Novas tecnologias emergentes
    - AplicaÃ§Ãµes prÃ¡ticas no mercado
    - Impacto nas empresas
    - Oportunidades de negÃ³cio
    
    ForneÃ§a informaÃ§Ãµes atualizadas e relevantes que possam ser usadas para criar conteÃºdo de marketing.
    """,
    agent=pesquisador,
    expected_output="RelatÃ³rio detalhado com as principais tendÃªncias de IA em 2024"
)

# Tarefa 2: AnÃ¡lise de Dados
tarefa_analise = Task(
    description="""
    Analise as informaÃ§Ãµes coletadas pelo pesquisador sobre tendÃªncias de IA.
    
    Identifique:
    - PadrÃµes e tendÃªncias principais
    - Oportunidades de mercado
    - Insights acionÃ¡veis para empresas
    - RecomendaÃ§Ãµes estratÃ©gicas
    
    Transforme os dados em insights prÃ¡ticos e acionÃ¡veis.
    """,
    agent=analista,
    expected_output="AnÃ¡lise detalhada com insights e recomendaÃ§Ãµes estratÃ©gicas",
    context=[tarefa_pesquisa]  # Depende da tarefa de pesquisa
)

# Tarefa 3: CriaÃ§Ã£o de ConteÃºdo
tarefa_conteudo = Task(
    description="""
    Crie um artigo de blog atrativo sobre as tendÃªncias de IA em 2024.
    
    Use os insights do analista para criar:
    - TÃ­tulo chamativo
    - IntroduÃ§Ã£o envolvente
    - ConteÃºdo estruturado e informativo
    - ConclusÃ£o com call-to-action
    
    O artigo deve ser interessante para profissionais de tecnologia e marketing.
    """,
    agent=redator,
    expected_output="Artigo completo sobre tendÃªncias de IA em 2024",
    context=[tarefa_analise]  # Depende da tarefa de anÃ¡lise
)

print("âœ… Todas as tarefas foram criadas com sucesso!")
print(f"   Tarefa 1: {tarefa_pesquisa.description[:50]}...")
print(f"   Tarefa 2: {tarefa_analise.description[:50]}...")
print(f"   Tarefa 3: {tarefa_conteudo.description[:50]}...")
print("=" * 60)
````

### **ðŸ‘¥ Formando Nossa Primeira Equipe (Crew)**

Agora vamos juntar nossos agentes e tarefas para formar nossa primeira equipe! Ã‰ como montar o time que vai trabalhar no projeto.

#### **ðŸ” O que Ã© uma Crew?**
Uma **Crew** Ã© a equipe formada pelos agentes que vÃ£o trabalhar juntos. Ã‰ como formar um time de projeto:
- **Agentes**: Quem vai trabalhar
- **Tarefas**: O que precisa ser feito
- **Processo**: Como vÃ£o trabalhar juntos
- **Objetivo**: O que queremos alcanÃ§ar

#### **ðŸ”„ Tipos de Processo:**
- **Sequential**: Um depois do outro (como uma linha de montagem)
- **Hierarchical**: Com hierarquia (gerente â†’ funcionÃ¡rios)
- **Collaborative**: Trabalhando juntos

Vamos usar **Sequential** para nosso primeiro exemplo, onde cada agente depende do trabalho do anterior.

````python
# ðŸ‘¥ FORMANDO NOSSA PRIMEIRA EQUIPE (CREW)
# Vamos juntar nossos agentes e tarefas para trabalhar juntos

print("ðŸ‘¥ Formando nossa primeira equipe CrewAI...")

# Criando nossa crew
crew = Crew(
    agents=[pesquisador, analista, redator],  # Nossos agentes
    tasks=[tarefa_pesquisa, tarefa_analise, tarefa_conteudo],  # Nossas tarefas
    process=Process.sequential,  # Trabalhando em sequÃªncia
    verbose=True  # Mostra o que estÃ¡ acontecendo
)

print("âœ… Equipe formada com sucesso!")
print(f"   Agentes: {len(crew.agents)}")
print(f"   Tarefas: {len(crew.tasks)}")
print(f"   Processo: {crew.process}")
print("=" * 60)
````

Agora Ã© hora da verdade! Vamos executar nosso sistema e ver como nossos agentes trabalham juntos. Ã‰ como dar o "start" no projeto e ver a mÃ¡gica acontecer.

#### **ðŸŽ¯ O que vai acontecer:**
1. **Pesquisador** vai buscar informaÃ§Ãµes sobre IA
2. **Analista** vai processar e interpretar os dados
3. **Redator** vai criar o artigo final

#### **â±ï¸ Tempo estimado:**
Dependendo do LLM que vocÃª estÃ¡ usando, pode levar alguns minutos. Ã‰ normal!

````python
# ðŸš€ EXECUTANDO NOSSO PRIMEIRO SISTEMA CREWAI
# Vamos ver nossos agentes trabalhando juntos!

print("ðŸš€ Iniciando execuÃ§Ã£o do nosso sistema CrewAI...")
print("â±ï¸ Isso pode levar alguns minutos, dependendo do LLM...")
print("=" * 60)

try:
    # Executando nossa crew
    resultado = crew.kickoff()
    
    print("\nðŸŽ‰ SUCESSO! Nosso sistema CrewAI funcionou!")
    print("=" * 60)
    print("ðŸ“„ RESULTADO FINAL:")
    print("=" * 60)
    print(resultado)
    
except Exception as e:
    print(f"\nâŒ Erro durante a execuÃ§Ã£o: {e}")
    print("ðŸ’¡ Dica: Verifique se vocÃª configurou um LLM no setup inicial")
    
print("=" * 60)
````
